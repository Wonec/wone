```python
import requests
import re
```

```python
url = 'http://desk.tooopen.com/meinv.html'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36',
    'Referer':'http://imgfzone.tooopen.com/'
}
r = requests.get(url,headers=headers)
print(r.text)
```

```python
img_url = re.findall('img src="(.*?)"',r.text)
name = re.findall('<span>(.*?)</span>',r.text)
print(img_url)
print(name)
```

```python
num = 0
for img in img_url:
    response = requests.get(img,headers=headers)
    with open(f'{name[num]}.jpg','wb') as f:
        f.write(response.content)
    print(f'正在获取%d张图片,图片名称:{name[num]},图片地址:{img}'%num)
    num+=1
print('获取结束')
```
