# 折线图

```python
import numpy as np
import matplotlib.pyplot as  plt
```

```python
x = [5,6,7,11,19,25]
y1 = [1.5,1.8,1.0,2.4,2.0,3.5]
y2 = [3,4,4,3,2.5,2.8]
# 设置参数

plt.plot(x,y1,'s-',color='r')
plt.plot(x,y2,'o-',color='g')
# 绘制图像 's-''o-'代表折线位置圆点方点 color设置yan'b's

plt.xlabel('X轴')
plt.ylabel('Y轴')
# 设置XY轴名称

plt.show() # 显示图像
```

# 散点图

```python
import numpy as np
import matplotlib.pyplot as  plt
```

```python
n = 1000
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
# 设置1000个正太分布的随机数
plt.scatter(X,Y,alpha=0.5)
# 绘制图像 设置透明度0.5

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
# 设置XY轴区间

plt.show()
```

# 3D散点图

```python
import numpy as np
import matplotlib.pyplot as  plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
# 创建三维坐标

x = [1,2,3,4]
y = [3,4,5,6]
z = [6,7,8,9]
# 设置参数

ax.scatter(x,y,z)
# 绘制三维散点图
plt.show()
```
